```python
metadata = hh_adata.obs.copy()
```

```python
MGC1_meta = metadata[metadata["final_cluster"] == "Micro1"].copy()
MGC2_meta = metadata[metadata["final_cluster"] == "Micro2"].copy()
```

```python
prefix = "/tscc/projects/ps-epigen/users/biy022/scmethylhic/human_hippocampus/snm3c"
def build_path(sample, index):
    group = index.rsplit("-", 1)[0]
    path = "{}/{}_deep/fastq_demultiplex/{}/allc/{}.allc.tsv.gz".format(prefix, sample, group, index)
    return path
```

```python
MGC1_meta["AllcPath"] = MGC1_meta.apply(lambda r: build_path(r["sample"], r["index"]), axis=1)
MGC2_meta["AllcPath"] = MGC2_meta.apply(lambda r: build_path(r["sample"], r["index"]), axis=1)
```

```python
MGC1_meta["AllcPath"].to_csv("MGC_subtypes/MGC1_AllcPaths.tsv", sep="\t", header=False, index=True)
MGC2_meta["AllcPath"].to_csv("MGC_subtypes/MGC2_AllcPaths.tsv", sep="\t", header=False, index=True)
```
