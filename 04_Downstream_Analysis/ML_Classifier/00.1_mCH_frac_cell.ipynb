```python
import pandas as pd
import os
from pathlib import Path
```

```python
root_dir = Path("/tscc/projects/ps-epigen/users/biy022/scmethylhic/human_hippocampus/snm3c/hh_micro_mono/")
os.chdir(root_dir / "DMR_classifier")
```

```python
allc_meta = pd.read_csv("allc_meta.tsv", sep="\t", index_col=0, header=0)
```

```python
allc_meta[allc_meta["celltype"] == "Micro1"]["allc_path"][100]
```

```python
ct = "Mono2"
ct_meta = allc_meta[allc_meta["celltype"] == ct].copy()
ct_meta["mCHFrac"] = 0.0
```

```python
for cb in ct_meta.index:
    print("Processing {}...".format(cb))
    allc_path = ct_meta.loc[cb, "allc_path"]
    allc_df = pd.read_csv(allc_path, sep="\t", header=None)
    allc_df.columns = ["chrom", "position", "strand", "context", "mc", "cov", "methylated"]
    allc_df_mCH = allc_df[~allc_df["context"].str.startswith("CG")].copy()
    ct_meta.loc[cb, "mCHFrac"] = (allc_df_mCH["mc"] > 0).sum() / (allc_df_mCH["cov"] > 0).sum()
```

```python
ct_meta.to_csv("Mono2_meta.tsv", sep="\t", header=True, index=True)
```
